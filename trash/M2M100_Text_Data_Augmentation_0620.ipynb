```python
df.drop_duplicates(['comment'],inplace=True)
df
```

```python
list_star = list(df['star'])
list_comment = list(df['comment'])
```

```python
import re
```

```python
def preprocess_sentence(sent):
  sent = re.sub(r"([?.!,¿])", r" \1", sent)
  sent = re.sub(r"[^ㄱ-ㅎ가-힣!.?]+", r" ", sent)
  sent = re.sub(r"\s+", " ", sent)
  return sent
```

```python
new_df = { 'comment':[],
           'star':list_star}
for sent in list_comment:
  new_df['comment'].append(preprocess_sentence(sent))
```

```python
new_df = pd.DataFrame(new_df)
new_df
```

```python
new_df['label'] = 1
new_df
```

```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")

def create_model():

  model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")

  return model
```

```python
with strategy.scope():
  model = create_model()
```

```python
from tqdm.notebook import tqdm

new_dict1 = {'comment':[],
           'label':list_label}

for comment in tqdm(list_comment):

  tokenizer.src_lang = "ko"
  encoded_ko = tokenizer(comment, return_tensors="pt")
  generated_tokens = model.generate(**encoded_ko, forced_bos_token_id=tokenizer.get_lang_id("ko"))
  result = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]

  new_dict1['comment'].append(result)
```

```python
new_df1 = pd.DataFrame(new_dict1)
new_df1.head()
```
