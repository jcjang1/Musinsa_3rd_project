```python
def preprocess_sentence(sent):
    sent = re.sub(r"([?.!,¿])", r" \1", sent)
    sent = re.sub(r"[^ㄱ-ㅎ가-힣!.?]+", r" ", sent)
    sent = re.sub(r"\s+", " ", sent)
    return sent
```

```python
a = a['comment'].apply(str).apply(preprocess_sentence)
```

```python
a = pd.DataFrame(a)
```

```python
a.drop_duplicates(['comment'], inplace=True)
a.reset_index(inplace=True)
```

```python
a.to_csv('fin_EDA_all_data_0621.csv', index=False)
```
