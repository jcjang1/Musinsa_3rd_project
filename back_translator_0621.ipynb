# Make back_translator function

````python
from tqdm.notebook import tqdm
from googletrans import Translator

def back_translator(sent, dests):
    translator = Translator()
    try:
        result_a = translator.translate(sent, dest=dests)
        result_b = translator.translate(result_a.text, dest='ko')
        result = result_b.text

        return result
    except:
        print('pass')
        pass
````

````python
commnet_dic = {
    'comment' : [],
    'star' : [],
    'language' : []
}
````

````python
for d in tqdm(['es']):
    for c, s in zip(tqdm(df['comment']), df['star']):
        commnet_dic['comment'].append(back_translator(c, d))
        commnet_dic['star'].append(s)
        commnet_dic['language'].append(d)
````

````python
dic = pd.DataFrame(commnet_dic)
dic
````

# Save data

````python
dic.drop_duplicates(['comment'],inplace=True)
dic.reset_index(inplace=True)
dic = dic[['comment', 'star']]
dic
````
