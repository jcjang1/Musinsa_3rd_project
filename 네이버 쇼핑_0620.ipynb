```python
from bs4 import BeautifulSoup
import lxml
import requests
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver import ActionChains
from fake_useragent import UserAgent
from selenium.webdriver.common.action_chains import ActionChains
import json 
from tqdm import tqdm
from time import sleep
import requests
import re
import pandas as pd
import numpy as np
import os

from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome("./chromedriver")   
test=[]

driver.get("https://swindow.naver.com/style/fashionbrand/list/category")   ### 크롤링 사이트 진입
time.sleep(2)

num = 2000

for i in range(0, num // 20):
    action = ActionChains(driver)
    scroll=driver.find_element_by_xpath(f'//*[@id="grid"]/li[{19 + (i*20)}]')
    action.move_to_element(scroll).perform()
    time.sleep(3)

for i in tqdm(range(1, num+1)):
    a=driver.find_element_by_xpath(f'//*[@id="grid"]/li[{i}]/a[1]')  # xpath 붙여넣기  [@id="btfTab"] ->[@id='btfTab'] 
    a=a.get_attribute('href')
    test.append(a)
    time.sleep(3)
```
